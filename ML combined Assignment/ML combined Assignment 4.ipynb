# Problem Statement:
# In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = sns.load_dataset('iris')
```

```python
t = decomposition.PCA(n_components=3).fit_transform(x.drop(['species'],axis=1))
x_new = pd.DataFrame(t)
x_new['species'] = x['species']
```

```python
t = x_new['species'].map({'setosa':0,'virginica':1,'versicolor':2})

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x_new[0],x_new[1],x_new[2],c=t,s=40)
```
