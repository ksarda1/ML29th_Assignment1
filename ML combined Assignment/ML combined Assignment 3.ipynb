# Problem Statement :
# In this assignment students have to compress racoon grey scale image into 5 clusters. In the end, visualize both raw and compressed image and look for quality difference.

```python
# importing the required libraries

from scipy.misc import face
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# reading the data
data = face()

# reshaping the data
data_r = data.reshape(-1,3)
```

```python
# implementing the KMeans by fitting the image data

kmeans = KMeans(n_clusters=5)
kmeans.fit(data_r)
```

```python
# replcaing the data with its cluster center

data_t = kmeans.cluster_centers_[kmeans.labels_]
```

```python
# reshapping tha data for plotting tha image

data_t = data_t.reshape(len(data),len(data[0]),3)
```

```python
# rounding the pixel value from 0 to 1 since cluster center is float value

data_t = data_t/255
```

```python
# plotting the image after clustering and image before clustering

plt.imshow(data_t)
plt.show()
plt.imshow(data)
plt.show()
```

So clustered image is less colored but we can easily understand the content
